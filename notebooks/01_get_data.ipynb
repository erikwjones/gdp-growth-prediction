```python
import numpy as np
import pandas as pd

from fredapi import Fred
import os
```

```python
fred = Fred(api_key=os.getenv("FRED_API_KEY"))
```

```python
start = "1980-01-01"
end = "2024-01-01"

fred_series = {
    "gdp": "GDPC1",
    "unemployment": "UNRATE",
    "cpi": "CPIAUCSL",
    "fed_funds": "FEDFUNDS",
    "ten_year": "GS10",
    "three_month": "TB3MS",
}

data = pd.DataFrame()

for name, code in fred_series.items():
    data[name] = fred.get_series(code, start, end)

data.index = pd.to_datetime(data.index)
```

```python
data["gdp_growth"] = np.log(data["gdp"]).diff()
data["yield_curve"] = data["ten_year"] - data["three_month"]
```

```python
data = data[["gdp_growth", "unemployment", "cpi", "fed_funds", "yield_curve"]]
# drop first row due to diff()
data = data.dropna()
```

```python
data.to_csv("../data/cleaned_data.csv")
```
